```python
import pandas as pd
import numpy as np
from typing import Tuple
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.datasets import make_classification
```

```python

X, y = make_classification(n_samples=1000, n_features=10, n_informative=5, random_state=42)


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(f"Training data shape: {X_train.shape}")
print(f"Testing data shape: {X_test.shape}")


feature_engineering_model = Pipeline([
    ('scaler', StandardScaler()),        # Feature engineering step: Standardize features
    ('classifier', LogisticRegression()) # Modeling step: Logistic Regression classifier
])

# 4. Train the pipeline on the training data
print("\nTraining the model...")
feature_engineering_model.fit(X_train, y_train)
print("Model training complete.")


accuracy = feature_engineering_model.score(X_test, y_test)
print(f"\nModel accuracy on the test set: {accuracy:.4f}")
```
